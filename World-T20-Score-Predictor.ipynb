```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
df = pickle.load(open("Cleaned_t20_Dataset.pkl", "rb"))
```

```python
df.head()
```

```python
df["city"].isna().sum()
```

```python
df["city"] = np.where(df["city"].isna(), df["venue"].str.split(" ").apply(lambda x: x[0]), df["city"])
```

```python
df["city"].isna().sum()
```

```python
df.head()
```

```python
df.drop(columns = ["venue"], inplace = True)
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df.head()
```

```python
df["current_score"] = df.groupby("match_id")["runs"].cumsum()
```

```python
df.head()
```

```python
df["over"] = df["ball"].astype(str).str.split(".").apply(lambda x: x[0]).astype(int)
df["ball_no"] = df["ball"].astype(str).str.split(".").apply(lambda x: x[1]).astype(int)
```

```python
df.head(10)
```

```python
df["balls_bowled"] = df["over"]*6 + df["ball_no"]
```

```python
df
```

```python
df["balls_left"] = 120 - df["balls_bowled"]
df["balls_left"] = df["balls_left"].apply(lambda x: 0 if x < 0 else x)
```

```python
df
```

```python
df["player_dismissed"] = df["player_dismissed"].apply(lambda x: 0 if x == "0" else 1)
df["player_dismissed"] = df["player_dismissed"].astype(int)
df["player_dismissed"] = df.groupby("match_id")["player_dismissed"].cumsum()
df["wickets_left"] = 10 - df["player_dismissed"]
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df.head()
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df["last_five"] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id')["runs"].sum().reset_index().merge(df,on='match_id')
```

```python
final_df
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
final_df.dropna(inplace = True)
```

```python
final_df.isnull().sum()
```

```python
final_df.shape
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.head()
```

## Train Test Split

```python
x = final_df.drop(columns= ["runs_x"])
y = final_df["runs_x"]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 1)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
```

## XGBRegressor

```python
trf = ColumnTransformer(transformers=[
    ("trf", OneHotEncoder(drop = "first", sparse = False), [0, 1, 2])
], remainder = "passthrough")

pipe = Pipeline([
    ("trf1", trf), 
    ("trf2", StandardScaler()),
    ("trf3", XGBRegressor(n_estimators = 1000, learning_rate = 0.2, max_depth = 12, random_state = 1))
])

pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)

print("R2 Score: -", r2_score(y_test, y_pred))
print("MAE: -", mean_absolute_error(y_test, y_pred))
```

## Save the model

```python
pickle.dump(pipe, open("model.pkl", "wb"))
```
